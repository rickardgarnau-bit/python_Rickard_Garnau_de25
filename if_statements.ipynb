## if-elif-else

```python
weight = 25

if weight < 20:
    print('Give patient 500mg')
elif weight < 40:
    print('Give 750mg')
else:
    print('Give 1000mg')
```

## nested if statements

```python
number = 25
if number % 5 == 0:
    print(f'{number} is divisible by 5')
    if number % 2 == 0:
        print(f'{number} is even')
    else:
        print(f'{number} is odd')
```
