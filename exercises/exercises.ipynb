# Glossary

### terminology	 explanation



git → A distributed version control system used to track changes in code/projects.

github → A cloud-based platform that hosts Git repositories, enabling collaboration, sharing, and remote storage.

version control → A system that keeps track of changes to files (especially code), so you can revert, compare, or collaborate.

backup → Keeping a safe copy of your code (Git acts like a smart backup system with history).

gitignore → A file that lists which files/folders Git should ignore (e.g., logs, temp files, secrets).

commit → A saved snapshot of your changes in the repository.

pull → Downloads (and merges) the latest changes from a remote repository (like GitHub) into your local repo.

clone → Copies an entire repository from a remote source (e.g., GitHub) to your computer.
