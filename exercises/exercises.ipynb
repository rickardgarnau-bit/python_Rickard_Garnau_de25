```python
tal1 = float(input('Välj ett tal'))
tal2 = float(input('Välj ett andra tal'))

summa = tal1 + tal2
skillnad = tal1 - tal2
produkt = tal1 * tal2
kvot = tal1 / tal2

print('Summan är: ', summa)
print('Skillnaden är: ', skillnad)
print('Produkten är: ', produkt)
print('Kvoten är :', kvot)
```
