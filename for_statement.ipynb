# for statement

```python
for i in range(3):
    print(i)
```

```python
for i in range(3):
    print(f'number {i}')
```

```python
for i in range(10):
    print(i, end = " ")
```

```python
for i in range(2, 10):
    print(i, end = " ")
```

```python
for i in range(2, 10, 2):
    print(i, end = " ")
```

```python
sum_ = 0 

for i in range(6):
    sum_ += i

print(f'0+1+2+3+4+5+ = {sum_}')
```

```python
import random as rnd

dice_rolls = 10_000
num_six = 0

for i in range(dice_rolls):
    dice = rnd.randint(1,6)
    if dice == 6:
        num_six += 1

print(f'Number of sixes {num_six} in {dice_rolls} dice rolls')
print(f'Probability of getting a six {num_six/dice_rolls}')
```

# loop trough a list

```python
fruits = ['apple', 'orange', 'pear', 'banana']
fruits
```

```python
print('I love to eat')
for fruit in fruits:
    print(fruit)
```

```python
print('My favorite fruits are')
for i, fruit in enumerate(fruits, 1):
    print(f'{i}. {fruit}')
```

```python
fruits = ['apple', 'orange', 'pear', 'banana']
recipes = ['pie', 'juice', 'cake', 'drink']

print(f"{'FRUIT':<20}RECIPE")
for fruit, recipe in zip(fruits, recipes):
    print(f'{fruit:<20}{recipe}')
```

# Nested for-loops

```python
for x in range(3):
    for y in range(3):
        print(f'{x},{y}', end = " ")
    print()
```
