# Matplotlib
- focus on the distinction between two ways of writing matplotlib codes
- in this course, focus on the components of a matplotlib figure

```python
import matplotlib.pyplot as plt
import numpy as np

```

##### The first approach - function/matlab style

```python
plt.figure()
plt.plot(x,f)
plt.title("sign(x)")
plt.xlabel("x")
plt.ylabel("f")
```

```python
# generate data
x  = np.linspace(-5, 5, 100)
x
```

```python
# generate data f
f = np.sin(x)
f
```

##### The second approach - OOP

```python
fig = plt.figure(facecolor='turquoise') # creata one figure
ax = plt.axes() # on the figure, create an axe
ax.plot(x, f) # on the axe. create a plot
ax.set(xlim=(-10,10), ylim=(-5,5), xlabel="x", ylabel="y") # continue modify the axe object in different ascpects
```
