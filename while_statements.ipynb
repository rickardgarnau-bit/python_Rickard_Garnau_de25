# while statements

```python
# Adding +1 to "n" using +=, else it continues forever. Now it stops at 5
n = 0
while n < 5:
    print(f'Hello {n}')
    n += 1
```

# Example - oil leakage
### There is an oil leakage causing the bird population in an island to 1/2 in each year. From start there were 80000 birds, how many yeas does it take for it to have 1/10 remaining?

```python
birds = 80000
year = 0

while birds > 8000:
    print(f'Year {year} there were {birds:.0f}s')
    birds /= 2
    year += 1

print(f'It takes {year} years for the birds to have less than 1/10th of population')
```

# break statement

```python
import random as rnd

while True:
    number1 = rnd.randint(1,10)
    number2 = rnd.randint(1,10)

    user_answer = int(input(f'What is {number1}*{number2}'))

    if user_answer == number1*number2:
        print('Correct')
    else:
        print(f'wrong, the answer was {number1*number2}')

    play_again = input('Wanna play again (y for yes)?')

    if play_again != 'y':
        print('Have a nice day!')
        break

                        
```
